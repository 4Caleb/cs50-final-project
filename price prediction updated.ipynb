```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

```python
df = pd.read_csv('nigeria-rent.csv')
df
```

```python
# Retrieving houses in Lagos and Abuja states only.

#since the last word in an adress is the state, we apply split() on last word in location column as follows 

df1 = df.copy()
df1['split'] = df1['Location'].apply(lambda x: x.split()[-1])


df1 = df1[(df1["split"] == 'Abuja') | (df1["split"] == 'Lagos')]
df1


    

    
```

```python
df2 = df1.copy()
```

```python
df2.Price.unique()
```

```python
df2["Title"] = df2["Title"].apply(str.lower)
df2
```

```python
#retrieving homes with a certain home type and removing commmercial properties
home_type = ['bungalow','duplex','terrace duplex','semi detached duplex','mansion','flat','apartment','detached duplex']
df3 = df2[df2["Title"].str.contains('|'.join(home_type))]
df3=df3[~df3['More Info'].str.contains('COMMERCIAL')]
df3
```

```python
#Flats and apartments are the same. Below code changes occurence of 'apartment' to flat for easier categorisation

df3['type'] = df3['Title'].apply(lambda x: 'flat' if 'apartment' in x else x)
```

```python
#arranging home types into 'Home' column
df3['Home'] = df3['type'].apply(lambda x: 'SEMI DETACHED DUPLEX' if 'semi detached duplex' in x else x)

for home in ['terrace duplex','detached duplex','mansion','duplex','bungalow','flat']:

    df3['Home'] = df3['Home'].apply(lambda x: f'{home}'.upper() if home in x else x)
 
df3
```

```python
#making dummy variables for Abuja and Lagos states
df3['Abuja'] = df3['split'].apply(lambda x: int('Abuja' in x))
```

```python
#removing unnecesary data
df4 = df3.copy()
index1 = df4[(df4['Bedrooms']==' beds') | (df4['Bedrooms'] == '0 beds')].index


df4 = df4.drop(index1)
index2 =  df4[(df4['Toilets']==' Toilets') | (df4['Toilets'] == '0 Toilets')].index

df4 = df4.drop(index2)
index3 = df4[(df4['Bathrooms']==' baths') | (df4['Bathrooms'] == '0 baths')].index

df4 = df4.drop(index3)


```

```python
#changing bedrooms, bathrooms and toilets to int type
rooms = ['Bedrooms','Bathrooms','Toilets']

for room in rooms:
     df4[f'{room}'] = df4[f'{room}'].apply(lambda x: int(x[0]))
    

# Turning price column to int type
df4['Price'] = df4['Price'].apply(lambda x: int(x.split('/')[0].replace(',','')))


                                        
```

```python
#filtering price extremes

df4 = df4[(df4['Price']<1700000000000) & (df4['Price']>100000)]

index1 = df4[(df4['Price']<=100000) & (df4['Bedrooms']>=3)].index
df4 = df4.drop(index1)
index2 = df4[(df4['Abuja'] == 0) &(df4['Price']==2.500000e+10)].index
df4 = df4.drop(index2)
```
